**LoanStatus**


How are loan status distributed?

```python
fig = plt.figure()
ax = fig.add_axes([.0, .0, 2, 1])
sb.countplot(data=df, x='LoanStatus').set_title("Loan status distribution");
ax.set_ylabel('Count of load status')
ax.set_xlabel('Load status')
plt.xticks(rotation=15);
```

**LoanStatus vs BorrowerState:**

Are there relationship between loan status and borrower state? Are there states that have unique behaviour of load status?
 

```python
fig = plt.figure()
ax = fig.add_axes([.0, .0, 2, 0.8])
ct_counts = df.groupby(['LoanStatus', 'BorrowerState']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index = 'LoanStatus', columns = 'BorrowerState', values = 'count')

sb.heatmap(ct_counts).set_title("LoanStatus vs BorrowerState")
```

**Observation:** 

We have the only few coloured blocks are
 - Current + CA
 - Completed + CA
 - Current + FL
 - Current + NY
 - Current + TX
 
We also notice that, 1/8 of data are CA and 50% of the data are current.
We take a closer look to past due section, since they are all black.

```python
df_group_pastDue = df.copy()
df_group_pastDue.loc[df_group_pastDue['LoanStatus'].str[:4] == 'Past', 'LoanStatus'] = 'Past Due'
```

Since past due data are all black, we exploration more with group them together.

```python
fig = plt.figure()
ax = fig.add_axes([.0, .0, 2, 0.5])
ct_counts = df_group_pastDue.groupby(['LoanStatus', 'BorrowerState']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index = 'LoanStatus', columns = 'BorrowerState', values = 'count')

sb.heatmap(ct_counts).set_title("LoanStatus vs BorrowerState")
```

**Observation:** Result does not change.

```python
df_pastDue = df.copy()

df_pastDue.drop(df_pastDue[df_pastDue['LoanStatus'].str[:4] != 'Past'].index, inplace = True)
df_pastDue.loc[df_pastDue['LoanStatus'].str[:4] == 'Past', 'LoanStatus'] = 'Past Due'
```

We take a look at past due loan in each status with all loan in each status order.

```python
freq = df['BorrowerState'].value_counts()

pastDue_order = freq.index
fig = plt.figure()

fig = plt.figure()
ax = fig.add_axes([.0, .0, 2, 0.6])

sb.countplot(data=df_pastDue, x='BorrowerState', color=base_color, order=pastDue_order);
ax.set_xlabel('Borrower state')
ax.set_ylabel('Count of past due loan')
ax.set_title('Borrower state past due distribution')
```

**Observation:** There is no obvious relationship between borrowerState and LoanStatus. Mostly count are depending on population on loanstatus but not on borrowstate.

We see TX has a slightly lower past due loan rate,NY and FL have a slightly higher past due loan rate.

```python
df_Comp = df.copy()

df_Comp.drop(df_Comp[df_Comp['LoanStatus'].str[:4] != 'Comp'].index, inplace = True)
```

Also we take at complete loan in each status with all loan in each status order.

**BorrowerAPR vs BorrowerState**

What is the relationship between borrower APR and borrower state? We might also want to find the difference of big borrowerAPR and small borrowerAPR

```python
fig = plt.figure()
ax = fig.add_axes([.0, .0, 3, 1])
my_order = df.groupby(by=["BorrowerState"])["BorrowerAPR"].mean().iloc[::-1].index
ax.set_title('BorrowerState vs BorrowerAPR')
sb.boxplot(data=df, y='BorrowerAPR', x='BorrowerState', color=base_color,order=my_order)
```

**Obversation:** 
we have the largest amount of loan in CA, CA is not having the highest borrower APR.
The higher Borrower APR are from AR and AL and MS.
The lower Borrower APR are from ME and IA.


Is there any change if we group small borrower APR and larger borrower APR?

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 15), sharey=True)

df_APR_small = df[df['BorrowerAPR'] <= df['BorrowerAPR'].mean()]
df_APR_big = df[df['BorrowerAPR'] > df['BorrowerAPR'].mean()]

sb.boxplot(ax=axes[0],data=df_APR_small, x='BorrowerAPR', y='BorrowerState', color=base_color)
axes[0].set_title('Small BorrowerAPR vs BorrowerState')

sb.boxplot(ax=axes[1],data=df_APR_big, x='BorrowerAPR', y='BorrowerState', color=base_color)
axes[1].set_title('Larger BorrowerAPR vs BorrowerState')
```

**Obversation:** 
we notice that, among smaller BorrowerAPR  and larger BorrowerAPR, the highest and lowest states are different.

Among the larger Borrower APR, the highest is LA, lowest is ME.
Among the smaller Borrower APR, the highest is SD, lowest is VT.
